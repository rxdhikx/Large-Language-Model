```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer  
from sklearn.metrics.pairwise import cosine_similarity  
```

```python
#    Matrix design specifications
def plot_cosine_similarity_matrix(cosine_similarity_matrix, dataset1_labels, dataset2_labels):
    plt.figure(figsize=(10, 8))
    plt.imshow(cosine_similarity_matrix, cmap='viridis', aspect='auto')
    plt.colorbar(label='Cosine Similarity')
    plt.xlabel('Generated Answer', fontsize=14)
    plt.ylabel('Best Answer', fontsize=14)
    plt.title('Truthful QA - Cosine Similarity Matrix for First 20 rows', fontsize=16)
    plt.xticks(ticks=np.arange(len(dataset2_labels)), labels=dataset2_labels, rotation=45, ha='right')
    plt.yticks(ticks=np.arange(len(dataset1_labels)), labels=dataset1_labels, rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Read the dataset into a DataFrame & specify number of rows 
df = pd.read_csv('/home/rravindra0463@id.sdsu.edu/models/results_100_Llama/100_res_truthful_qa_separated3.csv', nrows=20)

# Assuming 'best answer' and 'Generated Answer' are the column names
best_answers = df['best_answer'].values.astype('U')
generated_answers = df['Generated Answer'].values.astype('U')

# Initialize CountVectorizer to convert text to matrix of token counts
vectorizer = CountVectorizer().fit(best_answers)

# Transform the text data to matrix of token counts
best_answers_matrix = vectorizer.transform(best_answers)
generated_answers_matrix = vectorizer.transform(generated_answers)

# Calculate cosine similarity between the matrices
cosine_similarities = cosine_similarity(best_answers_matrix, generated_answers_matrix)

# Assuming you have labels for the datasets 
dataset1_labels = [f'Best Answer {i}' for i in range(len(best_answers))]
dataset2_labels = [f'Generated Answer {i}' for i in range(len(generated_answers))]

# Plot the cosine similarity matrix
plot_cosine_similarity_matrix(cosine_similarities, dataset1_labels, dataset2_labels)
```
